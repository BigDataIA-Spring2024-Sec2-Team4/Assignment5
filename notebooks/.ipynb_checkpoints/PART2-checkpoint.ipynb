```python
import openai
import pandas as pd

openai.api_key = 'OPENAI_API_KEY'  # API key

# Function adapted for v1/chat/completions to generate questions
def generate_questions_from_outcome(outcome, num_questions):
    questions = []
    for _ in range(num_questions):
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a knowledgeable tutor tasked with creating CFA-level questions based on specific learning outcomes. Generate a question with four options (A, B, C, D), clearly indicating the correct answer."},
                {"role": "user", "content": f"Learning outcome: {outcome}"}
            ]
        )

        if response.choices:
            question_text = response.choices[0].message['content'].strip()
            if question_text:  # Ensure non-empty responses are added
                questions.append(question_text)

    return questions

# Load the CSV into a DataFrame
df = pd.read_csv('/content/filtered_data.csv')

question_bank = []

# For each topic, generate questions from each of its learning outcomes
for index, row in df.iterrows():
    if len(question_bank) >= 50:
        break  # Stop if we have already generated 50 questions

    learning_outcomes = row['Learning_Outcomes'].split(';')
    num_questions_per_outcome = max(1, 50 // (len(learning_outcomes) * 3))

    for outcome in learning_outcomes:
        if len(question_bank) >= 50:
            break
        questions = generate_questions_from_outcome(outcome, num_questions_per_outcome)
        question_bank.extend(questions[:num_questions_per_outcome])  # Ensuring to not to exceed per outcome

# Output the number of generated questions
print(f"Total questions generated: {len(question_bank)}")
for question in question_bank[:5]:  # Display examples
    print(question, "\n")
```

```python
import openai
import pandas as pd
from sentence_transformers import SentenceTransformer

openai.api_key = 'OPENAI_API_KEY'  # API key

def generate_questions_from_outcome(outcome, num_questions):
    question_answers = []  # Storing tuples of (question, answer)
    for _ in range(num_questions):
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a knowledgeable tutor tasked with creating CFA-level questions based on specific learning outcomes. Generate a question with four options (A, B, C, D), clearly indicating the correct answer."},
                {"role": "user", "content": f"Learning outcome: {outcome}"}
            ]
        )

        if response.choices:
            text = response.choices[0].message['content'].strip()
            if text:  # Non-empty responses are added
                # Placeholder for how you might separate the question from the answer
                question, answer = text.rsplit('\n', 1)
                question_answers.append((question, answer))

    return question_answers

# Load the CSV into a DataFrame
df = pd.read_csv('/content/filtered_data.csv')

question_answer_bank = []

for index, row in df.iterrows():
    if len(question_answer_bank) >= 50:
        break  # Stop if we have already generated 50 question-answer pairs

    learning_outcomes = row['Learning_Outcomes'].split(';')
    num_questions_per_outcome = max(1, 50 // (len(learning_outcomes) * 3))

    for outcome in learning_outcomes:
        if len(question_answer_bank) >= 50:
            break  # Check again in case we reached the limit during question generation
        qas = generate_questions_from_outcome(outcome, num_questions_per_outcome)
        question_answer_bank.extend(qas[:num_questions_per_outcome])  # Ensuring not to exceed per outcome

# Convert to DataFrame and save to CSV
qa_df = pd.DataFrame(question_answer_bank, columns=['Questions', 'Answers'])
qa_df.to_csv('/content/question_answer_bank.csv', index=False)

print(f"Total question-answer pairs generated: {len(question_answer_bank)}")


```

```python
from pinecone import Pinecone, ServerlessSpec

# Initialize Pinecone with the new approach
pc = Pinecone(
    api_key='PINECONE_KEY'
)

# Initialize your embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate vectors for your questions
question_vectors = model.encode(qa_df['Questions'].tolist())

# Generate vectors for your answers
answers_vectors = model.encode(qa_df['Answers'].tolist())


index_name = 'set-a'

# Check if the index exists, and if not, create it
if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=question_vectors.shape[1],
        metric='cosine',  # Or 'euclidean', based on your preference
        spec=ServerlessSpec(
            cloud='aws',  # Or 'gcp', based on your preference
            region='us-east-1'  # Choose the region closest to your users
        )
    )

# Connect to your index
pcIndex = pc.Index(index_name)

questions_to_insert=[]
answers_to_insert=[]
for index, (question, answer, question_vector, answer_vector) in enumerate(zip(qa_df['Questions'], qa_df['Answers'], question_vectors, answers_vectors), start=1):
    questions_to_insert.append((str(index), question_vector.tolist(), {"Question" : question}))
    answers_to_insert.append((str(index), answer_vector.tolist(), {"Answer" : answer}))

# Insert the data
pcIndex.upsert(vectors=questions_to_insert, namespace='Questions')

pcIndex.upsert(vectors=answers_to_insert, namespace='Answers')

print("Data inserted successfully!")



```
